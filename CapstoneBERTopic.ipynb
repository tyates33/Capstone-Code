```python
import pandas as pd
import numpy as np
```

```python
z_df = pd.read_csv("C:/Users/Tanner/OneDrive/Documents/School_Work/Capstone/2024-02-03_oth_export_ideas.csv")
```

```python
z_df['user_created_at'] = pd.to_datetime(z_df['user_created_at'])
z_df['user_created_at'] = z_df['user_created_at'].dt.date

z_df['idea_created_at'] = pd.to_datetime(z_df['idea_created_at'])
z_df['idea_created_at'] = z_df['idea_created_at'].dt.date
print(z_df.dtypes)
```

```python
#count number of null values in each column
print(z_df.isnull().sum())
```

```python
# drop columns where the idea description is null
z_df_clean = z_df.dropna(subset=['idea_description'])
print(z_df_clean.isnull().sum())
```

```python
#replace the null values in columns with object type with the string ' '
z_df_clean = z_df_clean.fillna(' ')
print(z_df_clean.isnull().sum())
```

```python
good_content_df = z_df_clean[['idea_description', 'canvas_problem', 'canvas_solution', 'canvas_unique_value_proposition', 'canvas_high_level_concept', 'canvas_channels', 'canvas_revenue_streams', 'canvas_cost_structure', 'canvas_unfair_advantage', 'canvas_existing_alternatives', 'competitors_direct', 'competitors_indirect', 'competitors_alternatives']]
```

```python
good_content_df['total'] = good_content_df['idea_description'] + good_content_df['canvas_problem'] + good_content_df['canvas_solution'] + good_content_df['canvas_unique_value_proposition'] + good_content_df['canvas_high_level_concept'] + good_content_df['canvas_channels'] + good_content_df['canvas_revenue_streams'] + good_content_df['canvas_cost_structure'] + good_content_df['canvas_unfair_advantage'] + good_content_df['canvas_existing_alternatives'] + good_content_df['competitors_direct'] + good_content_df['competitors_indirect'] + good_content_df['competitors_alternatives']
```

```python
# remove duplicate rows
good_content_df = good_content_df.drop_duplicates(subset=['total'])
```
